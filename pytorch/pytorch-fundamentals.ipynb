### Random Tensors

Why random tensors?

Random tensors are important because the way many neural networks learn is that they start with tensors full of random numbers and then adjust those numbers to better represents the data.

`Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers`

Torch random tensors - https://pytorch.org/docs/stable/generated/torch.rand.html

```python
# Create a random tensor of size (3, 4)
random_tensor = torch.rand(3, 4)
print(random_tensor)
random_tensor.ndim
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(224, 224, 3)) # height, width, colort channels (R, G, B)
random_image_size_tensor.shape, random_image_size_tensor.ndim
```

### Matrix Multipliction

Two main ways of performing multiplication in neural networks and deep learning:
1. Element-wise multiplication
2. Matrix multiplication (dot product)

More information on multiplying matrices - https://www.mathsisfun.com/algebra/matrix-multiplying.html


There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match:
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work

2. The resulting mateix has the shape of the **outer dimensions**:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
# Element wise multiplication 
print(tensor, '*', tensor)
print(f'Equals: {tensor * tensor}')
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

### One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 8],
                         [9, 10],
                         [11, 12]])

# torch.matmul(tensor_A, tensor_B)
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our tensor shape issues, we can manipulate the shape of one of our tensors using a **transpose**.

A **transpose** switches the axes or dimensions of a give tensor.

```python
tensor_B.T, tensor_B
```

```python
torch.matmul(tensor_A, tensor_B.T).shape
```

```python
# The matrix multiplication operation works when `tensor_B` transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape}, tensor_B.T = {tensor_B.T.shape}")
print(f"Muptiplying {tensor_A.shape}, @ {tensor_B.T.shape}")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

## Tensor Aggregation 

Finding the **min**, **max**, **mean**, **sum**, etc `torch.min()`, `torch.max()`, `torch.mean()`, `torch.sum()`

Finding the positional **min**, **max** `torch.argmin()`, `torch.argmax()`

```python
# Create a tensor
z = torch.arange(0, 100, 10)
z, z.dtype
```

```python
print(torch.min(z), z.min())
print(torch.max(z), z.max())
print(torch.mean(z.type(torch.float32)), z.type(torch.float32).mean()) # the `torch.mean()` requires a tensor of float32
print(torch.sum(z), z.sum())
```

```python
print(z.argmin(), z[0])
print(z.argmax(), z[9])
```

## Reshaping, stacking, squesing, and unsquesing tensor
* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeze - removes all `1` dimensions from a tensor
* Unsqueze - add a `1` dimensions to a target tensor
* Permute - return a view of the imput with dimensions permuted (swapped) in a certain way

```python
x = torch.arange(1., 10.)
print(x, x.shape)

# Add an extra dimensions
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
z[:, 0] = 5
z, x # changing the z changes x 
```

```python
# Stack tensors on top of each other
x0_stacked = torch.stack([x, x, x, x, x], dim=0)
x0_stacked, x0_stacked.shape
```

```python
x1_stacked = torch.stack([x, x, x, x, x], dim=1)
x1_stacked, x1_stacked.shape
```

```python
x_vstacked = torch.vstack([x, x, x, x, x])
x_vstacked, x_vstacked.shape
```

```python
x_hstacked = torch.hstack([x, x, x, x, x])
x_hstacked, x_hstacked.shape
```

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}\n")

# Removes all single dimensions from a target tensor
x_squeezed = x_reshaped.squeeze()
print(f"New tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}\n")

# Add a single dimensions to a target tensor
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"New tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
x_original = torch.rand(size=(224, 224, 3))

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute([2, 0, 1]) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_permuted.shape}")
print(f"New shape: {x_permuted.shape}")
```

```python
x_original[0, 0, 0] = 87622334
x_original[0, 0, 0], x_permuted[0, 0, 0]
```

## PyTorch tensors & Numpy

* Data in NumPy, want in PyTorch -> `torch.from_numpy(ndarray)`
* PyTorch tensor -> NumPy -> `torch.Tensor.numpy()`

```python
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# Change the value of array, what will this do to `tensor`?
array = array + 1
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Change the value of tensor, what will this do to `numpy_tensor`?
tensor = tensor + 1
tensor, numpy_tensor
```
